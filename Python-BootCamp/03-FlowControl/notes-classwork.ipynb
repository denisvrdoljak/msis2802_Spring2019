Write a program that takes a number as an input, computes the factorial of the number enterred, and then prints the result. *You must use a loop.* You may not use any libraries or math functions that calculate factorials.
* (Hint: a factorial is just the product of all numbers between 1 and the number. So, 5! = 5 * 4 * 3 * 2 * 1.)



### Problem 2, using a WHILE Loop

```python
i = 2000
answers = []
while i<= 3200:
    if not i%7 and i%5:
        answers.append(i)
    i +=1
print(answers)
        
```

### Problem 2, using a FOR Loop

```python
answers = []
for i in range(2000,3201):
    if not i%7 and i%5:
        answers.append(i)
print(answers)
```

```python
userinput = input("Enter a number.\n: ")
userinput = int(userinput)
answer = 1
for i in range(1,userinput+1):
    answer *= i
print(answer)
```

## Problem 3
Write a program that accepts an input, and then calculate the number of letters and digits in the entered string.
For example, this input:
hello world! 123
should return this output:
LETTERS 10
DIGITS 3

```python
lettercount = 0
numbercount = 0
userinput = input("Enter a string of numbers/letters.\n: ")
for letter in userinput:
    if letter.isalpha():
        lettercount += 1
    elif letter.isdigit():
        numbercount += 1
print("Letters: {}\nNumbers: {}".format(lettercount,numbercount))
```
